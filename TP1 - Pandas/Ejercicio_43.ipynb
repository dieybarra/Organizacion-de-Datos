```python
contents = pd.read_csv('contents.csv', usecols=['id', 'revision_id','revision_timestamp','revisor_username', 'revisor_ip'])
contents
```

```python
contents['revision_timestamp'] = pd.to_datetime(contents['revision_timestamp'])
```

```python
revisores = contents.groupby(['revisor_username']).count().reset_index().revisor_username.to_frame()
revisores
```

```python
revisores_ip = contents.groupby(['revisor_ip']).count().reset_index().revisor_ip.to_frame()
revisores_ip
```

```python
revisores_ip.value_counts()
```

```python
nombres = blocks.groupby(['title']).agg({'timestamp':'min'}).reset_index()
nombres
```

```python
#Me fijo si en 'title' los nombres siempre empiezan con 'Usuario:'
prueba = nombres['title'].str[0:8]=='Usuario:'
prueba.value_counts()
```

```python
nombres['title'] = nombres['title'].str[8:]
nombres
```

```python
revisores_bloqueados = revisores.merge(nombres, 'inner', left_on= 'revisor_username', right_on= 'title')[['title', 'timestamp']]
revisores_bloqueados
```

```python
revisores_ip_bloqueados = revisores_ip.merge(nombres, 'inner', left_on= 'revisor_ip', right_on= 'title')[['title', 'timestamp']]
revisores_ip_bloqueados
```

```python
resultado_a = len(revisores_bloqueados) + len(revisores_ip_bloqueados)
resultado_a
```

2da Parte

```python
merged = contents.merge(nombres, 'inner', left_on= 'revisor_username', right_on= 'title')
merged
```

```python
merged['revision_timeint'] = merged['revision_timestamp'].astype(np.int64)
merged
```

```python
merged = merged.groupby('revisor_username')['revision_timeint', 'timestamp'].agg({'revision_timeint':'mean', 'timestamp':'min'}).reset_index()
merged
```

```python
merged['revision_mean'] = pd.to_datetime(merged['revision_timeint'])
merged
```

```python
merged['diferencia'] = abs(pd.to_datetime(merged['timestamp'].astype(np.int64)) - merged['revision_mean'])
merged
```

```python
resultado = merged[['revisor_username', 'diferencia']]
resultado
```
